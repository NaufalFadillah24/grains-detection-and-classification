```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

file_path = r"C:\Users\Naufal\OneDrive\MyDocuments\KULIAH\Tugas Kuliah\TUGAS AKHIR\dataset\dataset hsv\numbered_gabah_features_completed.xlsx"
df = pd.read_excel(file_path)

# 2. Pilih kolom numerik yang ingin dinormalisasi
kolom_numerik = df.select_dtypes(include=['int64', 'float64']).columns

# 3. Buat objek scaler
scaler = MinMaxScaler()

# 4. Lakukan normalisasi
df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])

# 5. Simpan dataset yang sudah dinormalisasi
df.to_excel('normalized_numbered_gabah_features_completed2.xlsx', index=False)

print("Normalisasi selesai dan disimpan ke 'normalized_numbered_gabah_features_completed.xlsx'")
```
